```python
df.groupby(['multiMovieYn', 'repNationCd'], dropna=False, observed=False) \
.size() \
.reset_index(name='count')

df_grouped_movienm = df.groupby(['movieNm', 'multiMovieYn', 'repNationCd'], dropna=False, observed=False) \
                  .size() \
                  .reset_index(name='count')

df_grouped_nm = pd.DataFrame(df_grouped_movienm)

# 각 조건에 따른 boolean 컬럼 생성
df_grouped_nm['multiMovieYn_Y'] = df_grouped_nm['multiMovieYn'] == 'Y'
df_grouped_nm['multiMovieYn_N'] = df_grouped_nm['multiMovieYn'] == 'N'
df_grouped_nm['repNationCd_K'] = df_grouped_nm['repNationCd'] == 'K'
df_grouped_nm['repNationCd_F'] = df_grouped_nm['repNationCd'] == 'F'

# 생성된 boolean 컬럼들을 기준으로 그룹화하고 count
grouped_df = df_grouped_nm.groupby(['multiMovieYn_Y', 'multiMovieYn_N', 'repNationCd_K', 'repNationCd_F'])['count'].sum().reset_index()

print(grouped_df)
```

```python
# 'movieNm' 컬럼을 포함하여 그룹화하고 count
grouped_df_with_movienm = df_grouped_nm.groupby(['movieNm', 'multiMovieYn_Y', 'multiMovieYn_N', 'repNationCd_K', 'repNationCd_F'])['count'].sum().reset_index()

# 'movieNm' 별로 count 합계
grouped_by_movienm = grouped_df_with_movienm.groupby('movieNm')['count'].sum().reset_index()
grouped_by_movienm.rename(columns={'count': 'total_count'}, inplace=True)

print("movieNm별 count 합계:")
print(grouped_by_movienm)

# 'movieNm' 별로 각 boolean 조건 조합에 대한 count 합계
grouped_by_movienm_detailed = grouped_df_with_movienm.groupby('movieNm')[['multiMovieYn_Y', 'multiMovieYn_N', 'repNationCd_K', 'repNationCd_F', 'count']].sum().reset_index()

print("\nmovieNm별 boolean 조건 조합에 대한 count 합계:")
print(grouped_by_movienm_detailed)
```

```python
# 각 조건별 count를 위한 함수 정의
def get_counts(group):
    counts = pd.Series({
        'multiMovieYn=Y': group[group['multiMovieYn'] == 'Y']['count'].sum(),
        'multiMovieYn=N': group[group['multiMovieYn'] == 'N']['count'].sum(),
        'repNationCd=K': group[group['repNationCd'] == 'K']['count'].sum(),
        'repNationCd=F': group[group['repNationCd'] == 'F']['count'].sum()
    })
    return counts

# 각 movieNm별 count 적용 및 결과 병합
result_df = grouped_by_movienm.apply(get_counts).reset_index()

print(result_df)
```
