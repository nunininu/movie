```python
import pandas as pd
PATH = "~/data/movies/dailyboxoffice/dt=20240101"
df = pd.read_parquet(PATH)
len(df)
```

```python
df_grouped_nm = df.groupby(['movieNm', 'multiMovieYn', 'repNationCd'], dropna=False, observed=False) \
                  .size() \
                  .reset_index(name='count')

group = df_grouped_nm

print(df_grouped_nm)
```

```python
def get_counts(group): # 인자 이름을 group으로 변경 (더 명확하게 그룹 데이터임을 나타냄)
    counts = pd.Series({
        'multiMovieYn=Y': group[group['multiMovieYn'] == 'Y']['count'].sum() if 'count' in group.columns and 'multiMovieYn' in group.columns else 0,
        'multiMovieYn=N': group[group['multiMovieYn'] == 'N']['count'].sum() if 'count' in group.columns and 'multiMovieYn' in group.columns else 0,
        'repNationCd=K': group[group['repNationCd'] == 'K']['count'].sum() if 'count' in group.columns and 'repNationCd' in group.columns else 0,
        'repNationCd=F': group[group['repNationCd'] == 'F']['count'].sum() if 'count' in group.columns and 'repNationCd' in group.columns else 0
    })
    return counts
```

```python
# 각 movieNm별 count 적용 및 결과 병합
result_df = group.apply(get_counts).reset_index()

print(result_df)
```

```python
import pandas as pd

PATH = "~/data/movies/dailyboxoffice/dt=20240101"
df = pd.read_parquet(PATH)

# multiMovieYn, repNationCd과 추가로 movieNm를 group (movieNm는 추후 고유 키값으로 활용됨)
df_grouped_nm = df.groupby(['movieNm', 'multiMovieYn', 'repNationCd'], dropna=False, observed=False) \
                  .size() \
                  .reset_index(name='count')

# multiMovieYn, repNationCd 값별 count를 조건에 맞게 sumif
def get_counts(group):
    counts = pd.Series({
        'multiMovieYn=Y': group[group['multiMovieYn'] == 'Y']['count'].sum() if 'count' in group.columns and 'multiMovieYn' in group.columns else 0,
        'multiMovieYn=N': group[group['multiMovieYn'] == 'N']['count'].sum() if 'count' in group.columns and 'multiMovieYn' in group.columns else 0,
        'repNationCd=K': group[group['repNationCd'] == 'K']['count'].sum() if 'count' in group.columns and 'repNationCd' in group.columns else 0,
        'repNationCd=F': group[group['repNationCd'] == 'F']['count'].sum() if 'count' in group.columns and 'repNationCd' in group.columns else 0
    })
    return counts

# 각 movieNm별 count 적용 및 결과 병합
result_df = df_grouped_nm.groupby('movieNm', group_keys=False).apply(get_counts, include_groups=False).reset_index()

result_df #jupyter lab 
print(result_df) # print 
```
